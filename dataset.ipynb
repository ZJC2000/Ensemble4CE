```python
import os, glob
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

```python
samples = glob.glob(os.path.join('dataset/MAFWCE', '*', '*'))
vidlength = {clip[-5:]: len(os.listdir(clip)) for clip in samples}

audlength = {}
for aud in glob.glob('abaw/models--facebook--hubert-base-ls960-FRA/*'):
    audlength[aud[-9:-4]] = len(np.load(aud))
    # print(audlength)

#3s, 3s
```

```python
vl, al = [], []
for k in vidlength.keys():
    # print(vidlength[k], audlength[k])
    vl.append(vidlength[k])
    al.append(audlength[k])

```

```python
plt.scatter(Counter(vl).keys(), Counter(vl).values(), s=5)
plt.scatter(Counter(al).keys(), Counter(al).values(), s=5)
vl = dict(Counter(vl))
sorted(vl.items(), key= lambda x: x[1])
```
